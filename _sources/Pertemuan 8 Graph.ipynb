```python
import networkx as nx
G=nx.Graph()

G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(4,5),(4,6),(5,6),(5,7),(5,8),
                  (6,7),(6,8),(7,8),(7,9)])
```

```python
G.number_of_nodes()
```

```python
G.nodes()
```

```python
G.edges()
```

```python
G.neighbors(1)
```

```python
H=nx.DiGraph(G) # membuat digraph(agar menjadi 2 arah) dari G

H.edges()
```

```python
nx.draw(G, with_labels=True)
#nx.draw_random(G)
#nx.draw_circular(G)
#nx.draw_spectral(G)
```

```python
# Menghitung Degree Centrality
nx.degree_centrality(G)
```

```python
#Menghitung Closeness Centrality
nx.closeness_centrality(G)
```

```python
#Menghitung Betweenness Centrality
nx.betweenness_centrality(G)
```
