```python
from bs4 import BeautifulSoup
import requests
URL = "https://pta.trunojoyo.ac.id/c_search/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
results = soup.find(id="content")
#print(results.prettify())

karya_ilmiah_elements = results.find_all(attrs={"data-id": True})

for karya_ilmiah_element in karya_ilmiah_elements:
  judul_element = karya_ilmiah_element.find("a", class_="title")
  abstrak_element = karya_ilmiah_element.find("p")
  print("Judul :",judul_element.text.strip())
  print("Abstrak :",abstrak_element.text.strip())
  print()
```
