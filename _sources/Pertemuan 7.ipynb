```python
# transform to a supervised learning problem
X = split_sequence(label, 3)
# show each sample
for i in range(len(X)):
  print(X[i])
```

```python
df = pd.DataFrame(X, columns=['Input1', 'Input2', 'Output'])

path = '/content/drive/MyDrive/prosaindata/tugas/timeSeriesPassanger.csv'
with open(path, 'w', encoding = 'utf-8-sig') as f:
  df.to_csv(f)

ts = pd.read_csv('/content/drive/MyDrive/prosaindata/tugas/timeSeriesPassanger.csv')
ts

```
